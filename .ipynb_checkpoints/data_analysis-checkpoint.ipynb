```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
# loading the file into a pandas dataframe
df = pd.read_csv('clean_data.csv')
df.head()
```

```python
# Calculating the domestic/worldwide profit and putting them in new columns
df['domestic_profit'] = df['domestic_gross'] - df['production_budget']
df['worldwide_profit'] = df['worldwide_gross'] - df['production_budget']

df.head()
```

```python
# calculating the mean profit for each month
domestic_profit = df.groupby('TN_month')['domestic_profit'].mean()
worldwide_profit = df.groupby('TN_month')['worldwide_profit'].mean()

# Plotting the Release Months against their Mean Profit (Both worldwide and Domestic)
fig, ax = plt.subplots(figsize=(20, 10))

x = domestic_profit.index
y1 = worldwide_profit.values
y2 = domestic_profit.values

ax.bar(x,y1)
ax.bar(x,y2)
ax.tick_params(axis='x', labelsize= 15)
ax.set_xlabel('Release Month', fontsize= 20)
ax.set_ylabel('Mean Profit',fontsize= 20)
plt.suptitle('MEAN PROFIT BY RELEASE MONTH', fontsize= 25)

ax.legend(['Worldwide', 'Domestic'], fontsize= 20)
ax.set_ylim(0);
```

```python
# subsetting these columns and assigning to new variable
genre_profit = df.loc[:, ['genres', 'domestic_profit', 'worldwide_profit']]


# exploding the genres column
genre_profit_new = genre_profit.set_index(genre_profit.columns.drop('genres',1).tolist()).genres.str.split(',', expand=True).stack().reset_index().rename(columns={0:'genres'}).loc[:, genre_profit.columns]
genre_profit_new
```

```python
# calculating the mean profit for each genre
genre_domestic = genre_profit_new.groupby('genres')['domestic_profit'].mean()
genre_worldwide = genre_profit_new.groupby('genres')['worldwide_profit'].mean()

# Plotting the Genres against their Mean Profit (Both worldwide and Domestic)
fig, ax = plt.subplots(figsize=(20, 10))

x = genre_domestic.index
y1 = genre_worldwide.values
y2 = genre_domestic.values

ax.bar(x,y1)
ax.bar(x,y2)
ax.tick_params(axis='x', labelsize= 15, rotation=45)
ax.set_xlabel('Genres', fontsize= 20)
ax.set_ylabel('Mean Profit',fontsize= 20)
plt.suptitle('MEAN PROFIT BY GENRE', fontsize= 25)

ax.legend(['Worldwide', 'Domestic'], fontsize= 20)
ax.set_ylim(0);
```

```python
# subsetting these columns and assigning to new variable
directors_profit = df.loc[:, ['directors', 'domestic_profit', 'worldwide_profit']]

# exploding the directors column
directors_profit_new = directors_profit.set_index(directors_profit.columns.drop('directors',1).tolist()).directors.str.split(',', expand=True).stack().reset_index().rename(columns={0:'directors'}).loc[:, directors_profit.columns]
directors_profit_new
```

```python
# calculating the mean domestic profit for each associated director and selecting the top 50
directors_domestic = directors_profit_new.groupby('directors')['domestic_profit'].mean().sort_values(ascending=False)[:50]


# Plotting the directors against their Mean Domestic Profit
fig, ax = plt.subplots(figsize=(20, 10))

x = directors_domestic.index
y = directors_domestic.values

ax.bar(x,y)
ax.tick_params(axis='x', labelsize= 15, rotation=90)
ax.set_xlabel('Directors', fontsize= 20)
ax.set_ylabel('Mean Domestic Profit',fontsize= 20)
plt.suptitle('DIRECTORS ASSOCIATED WITH THE TOP 50 DOMESTIC MEAN PROFIT', fontsize= 25);
```

```python
# calculating the mean worldwide profit for each associated director and selecting the top 50
directors_worldwide = directors_profit_new.groupby('directors')['worldwide_profit'].mean().sort_values(ascending=False)[:50]


# Plotting the directors against their Mean Wordlwide Profit
fig, ax = plt.subplots(figsize=(20, 10))

x = directors_worldwide.index
y = directors_worldwide.values

ax.bar(x,y)
ax.tick_params(axis='x', labelsize= 15, rotation=90)
ax.set_xlabel('Directors', fontsize= 20)
ax.set_ylabel('Mean Worldwide Profit',fontsize= 20)
plt.suptitle('DIRECTORS ASSOCIATED WITH THE TOP 50 WORLDWIDE MEAN PROFIT', fontsize= 25);
```

```python
# subsetting these columns and assigning to new variable
genre_budget = df.loc[:, ['genres', 'production_budget']]

# exploding the genres column
genre_budget_new = genre_budget.set_index(genre_budget.columns.drop('genres',1).tolist()).genres.str.split(',', expand=True).stack().reset_index().rename(columns={0:'genres'}).loc[:, genre_budget.columns]
genre_budget_new
```

```python
# calculating the mean budget for each genre
genre_production = genre_budget_new.groupby('genres')['production_budget'].mean()

# Plotting the Genres against their Mean production budget 
fig, ax = plt.subplots(figsize=(20, 10))

x = genre_production.index
y = genre_production.values

ax.bar(x,y)
ax.tick_params(axis='x', labelsize= 15, rotation=90)
ax.set_xlabel('Genres', fontsize= 20)
ax.set_ylabel('Mean Budget',fontsize= 20)
plt.suptitle('MEAN PRODUCTION BUDGET PER GENRE', fontsize= 25);
```
